#### Use a box plot to display that same data.

```python
# Answer:
# create another column that sorts the rest of the data into bins
# plot box chart with bins as boxs 

def duration_bin(duration):
    if duration in range(0, 51):
        return '0 to 50'
    elif duration in range(50, 101):
        return '50 to 100'
    elif duration in range(100, 151):
        return '100 to 150'
    elif duration in range(150, 201):
        return '150 to 200'
    elif duration in range(200, 251):
        return '200 to 250'
    elif duration in range(250, 301):
        return '250 to 300'
    else:
        return '0'

movies['duration_tranch'] = movies.duration.apply(duration_bin)

movies.sort_values('duration', ascending = True)
```

```python
movies.boxplot(column=['duration'], by='duration_tranch');
```
